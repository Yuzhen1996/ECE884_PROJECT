```python
# Environment

NUM_MACHINES_SMALL = 5
NUM_TASKSPERJOB_SMALL = 5
NUM_JOBS_SMALL = 5

NUM_MACHINES_MEDIUM = 15
NUM_TASKSPERJOB_MEDIUM = 15
NUM_JOBS_MEDIUM = 15

NUM_MACHINES_LARGE = 20
NUM_TASKSPERJOB_LARGE = 20
NUM_JOBS_LARGE = 20

MIN_TASK_TIME = 1
MAX_TASK_TIME = 10

#SMALL 20 Descriptors
TASK_PROPERTIES_MACHINE_SMALL = [0, 3, 4, 2, 4, 4, 3, 1, 4, 3, 0, 3, 2, 2, 2, 3, 1, 1, 0, 1]
TASK_PROPERTIES_TIME_SMALL = [2, 5, 6, 6, 9, 4, 5, 6, 2, 7, 6, 3, 6, 2, 7, 1, 1, 9, 1, 6]

#MEIDUM 50 Descriptors
TASK_PROPERTIES_MACHINE_MEDIUM = [ 2, 10,  5,  0,  3,  6,  4, 11,  3, 10,  6,  5,  8, 12,  0,  1,  0, 4, 12,  1, 14, 11,  9,  5, 10, 13,  9, 11,  4,  4, 11, 11,  6, 13, 14, 13,  2,  0, 10,  5,  3, 11,  3,  2, 13, 10,  5,  3,  6,  3]
TASK_PROPERTIES_TIME_MEDIUM = [4, 9, 3, 5, 1, 1, 4, 1, 4, 5, 5, 3, 9, 3, 3, 4, 9, 8, 1, 6, 5, 7, 6, 1, 5, 4, 3, 6, 3, 3, 5, 6, 5, 4, 8, 4, 5, 8, 6, 8, 1, 3, 2, 7, 8, 5, 7, 2, 5, 6]

#LARGE 100 Descriptors
TASK_PROPERTIES_MACHINE_LARGE = [10, 12, 16,  5,  7, 16,  0, 15,  2, 12, 17,  3,  3,  9, 17, 11, 13,
        1, 19, 14,  2, 11, 11, 10,  7, 11, 17,  3,  2, 16, 18,  7, 19,  3,
       12, 10,  8, 11, 10,  5,  9, 12, 10,  3,  9,  5, 18,  8, 15, 16, 13,
       11,  9,  9, 11,  7, 18,  5, 18, 10,  4, 16, 15, 11, 14,  7, 17,  5,
        0, 10, 18, 17,  6, 10,  0,  8,  0,  8, 16, 10,  5, 14, 12,  0, 11,
       11,  3, 19,  9, 13, 13, 18, 10, 15, 10,  5,  2, 11,  0,  5]

TASK_PROPERTIES_TIME_LARGE = [3, 3, 1, 5, 4, 8, 9, 3, 3, 6, 4, 3, 2, 9, 8, 6, 7, 2, 2, 6, 4, 4,
       5, 2, 9, 4, 7, 7, 2, 2, 4, 4, 4, 4, 5, 7, 3, 8, 1, 4, 2, 6, 8, 5,
       1, 2, 1, 7, 4, 3, 8, 8, 8, 3, 5, 5, 9, 7, 5, 8, 5, 7, 9, 6, 3, 2,
       6, 9, 1, 5, 2, 6, 8, 2, 2, 1, 7, 2, 7, 1, 3, 4, 5, 5, 5, 2, 6, 5,
       4, 2, 8, 8, 8, 2, 3, 1, 5, 8, 6, 5]


class Config:
    
    def __init__(self):
    
        # Environment

        self.machine_profile = "small_default"
        self.job_profile = "small_default"
        self.task_profile = "small_default"  

        self.num_jobs = 0
        self.num_tasksperjob = 0
        self.num_machines = 0  
        
        self.task_properties_machine = []
        self.task_properties_time = []

        self.lambda_time = 1

        self._getMachinesProperties(self.machine_profile)
        self._getJobsProperties(self.job_profile)
        self._getTasksProperties(self.task_profile)

        # Network
        self.embedding_size = 64 # embedding size
        self.hidden_dim = 16     # agent LSTM num_neurons
        self.num_layers = 1      # agent LSTM num_stacks

        # Data
        self.batch_size = 128    # batch size


        # Training / test parameters
        self.num_epoch = 30000          # number of epochs
        self.learning_rate = 0.0001     # agent learning rate
        self.beta = 0                   # entropy regularization

        
        # Solver
        self.timeout = 1000

        
      
    def _getMachinesProperties(self, machine_profile):
       
        if machine_profile == "small_default":
            
            self.num_machines = NUM_MACHINES_SMALL
            
        elif machine_profile == "medium_default":

            self.num_machines = NUM_MACHINES_MEDIUM
            
        elif machine_profile == "large_default":

            self.num_machines = NUM_MACHINES_LARGE
            
        else:
            raise Exception('Machine profile not detected.')
            
    def _getJobsProperties(self, job_profile):
       
        if job_profile == "small_default":
            
            self.num_jobs = NUM_JOBS_SMALL
            self.num_tasksperjob = NUM_TASKSPERJOB_SMALL
            
        elif job_profile == "medium_default":

            self.num_jobs = NUM_JOBS_MEDIUM
            self.num_tasksperjob = NUM_TASKSPERJOB_MEDIUM
            
        elif job_profile == "large_default":

            self.num_jobs = NUM_JOBS_LARGE
            self.num_tasksperjob = NUM_TASKSPERJOB_LARGE
            
        else:
            raise Exception('Job profile not detected.')
            
    def _getTasksProperties(self, task_profile):

        if task_profile == "small_default":

            assert len(TASK_PROPERTIES_MACHINE_SMALL) == len(TASK_PROPERTIES_TIME_SMALL)
            
            self.task_properties_machine = TASK_PROPERTIES_MACHINE_SMALL
            self.task_properties_time = TASK_PROPERTIES_TIME_SMALL
            
        elif task_profile == "medium_default":

            assert len(TASK_PROPERTIES_MACHINE_MEDIUM) == len(TASK_PROPERTIES_TIME_MEDIUM)
            
            self.task_properties_machine = TASK_PROPERTIES_MACHINE_MEDIUM
            self.task_properties_time = TASK_PROPERTIES_TIME_MEDIUM

        elif task_profile == "large_default":

            assert len(TASK_PROPERTIES_MACHINE_LARGE) == len(TASK_PROPERTIES_TIME_LARGE)
            
            self.task_properties_machine = TASK_PROPERTIES_MACHINE_LARGE
            self.task_properties_time = TASK_PROPERTIES_TIME_LARGE
            
        else:
            raise Exception('Task profile not detected.')
            
    def reconfigure(self):
        
        self._getMachinesProperties(self.machine_profile)
        self._getJobsProperties(self.job_profile)
        self._getTasksProperties(self.task_profile)
```
